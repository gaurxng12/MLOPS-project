```python
import pandas as pd
import pymongo

df = pd.read_csv('data.csv')
df.head()
```

```python
# df should be converted into dict before we push it to mongodb
# df.to_dict(orient='records') converts the DataFrame into a list of dictionaries, where each row becomes one dictionary.
# orient='records' means:
# ➡️ Convert each row of the DataFrame into a separate dictionary.
# ➡️ Return a list of dictionaries, one dict per row.
data = df.to_dict(orient='records')
```

```python
from urllib.parse import quote_plus

DB_NAME = "Proj1"
COLLECTION_NAME = "Proj1-Data"
username = quote_plus("username")
password = quote_plus("password")
CONNECTION_URL = f"mongodb+srv://{username}:{password}@cluster0.wloqh7l.mongodb.net/?appName=Cluster0"

# above, either remove your credentials or delete the mongoDB resource bofore pushing it to github.
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
# Use the above defined variables to establish a connection with mongoDB setup
```

```python
# Uploading data to MongoDB
rec = collection.insert_many(data)
```
